```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('california_cities.csv')
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
cities
```

```python
#Extract Latd = Vi Do and LongD = Kinh Do
lat, lon = cities['latd'], cities['longd']
```

```python
#Plot using Pylot API
population = cities['population_total']
area = cities['area_total_km2']
plt.scatter(lon, lat, c = np.log10(population), cmap = 'viridis',
            s = area, linewidth = 0, alpha = 0.3);
plt.axis('equal');
plt.xlabel('longlatitude');
plt.ylabel('latitude');
plt.colorbar(label = 'log$_{10}$(population)');
plt.clim(3, 7)
plt.title('Califonia Cities: Population and Area Distribution')

# Create a legend for cities size
area_arange = [50, 100, 300, 500]
for area in area_arange:
    plt.scatter([], [], s = area, c = 'k', alpha = 0.2,
                label = str(area) + 'km$^2$')
plt.legend(labelspacing = 1, title = 'City Area');
```
