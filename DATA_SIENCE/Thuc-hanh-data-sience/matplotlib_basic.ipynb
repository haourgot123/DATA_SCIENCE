```python
#Pylot API
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), color = 'blue', linestyle='dashed', label = 'sin(x)')
plt.plot(x, np.cos(x), color = 'red', label = 'cos(x)')
plt.title('Since(x) and Cosine(x)')
plt.xlabel('X')
plt.ylabel('F(x)')
plt.legend()
plt.axis('equal')
#plt.axis('tight')
#plt.axis([0, 4, -0.7, 0.6])
#plt.xlim([0, 4])
```

```python
# Object Oriented API
#Step1: Prepare Data
x = [1, 2, 3, 4]
y = [11, 22 , 43, 55]
#Step2: subplot
fig, ax = plt.subplots(figsize = (5, 5)) # Figure size = Width & Height of the plot
#Plot the data
ax.plot(x, y);
ax.set(title='A simple Plot',
       xlabel = 'X axis',
       ylabel = 'Y axis');

```

```python
# OO API
#Prepare new Data
np.random.seed(0)
colors = np.random.rand(100)
sizes = 1000 *np.random.rand(100)
x = np.random.randn(100)
y = np.random.randn(100)
fig, ax  = plt.subplots()
img1 = ax.scatter(x, y, c = colors, s = sizes, cmap = 'viridis', alpha=0.3)
fig.colorbar(img1);
```

```python
# BAR CHART
# Vertical chart
# Horizontal
```

```python
# Vertical chart
# Preparing a dictionary
drink_prices = {'Coca' : 10,
                'Pepsi' : 20,
                'Fanta': 15}
fig, ax = plt.subplots()
ax.bar(drink_prices.keys(), drink_prices.values())
ax.set(title = 'The prices of Soft_Drink', ylabel = 'Price($)')
```

```python
fig, ax = plt.subplots()
ax.barh(list(drink_prices.keys()), list(drink_prices.values()));
```

```python
# Histogram
#prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
student_height[:10]
```

```python
#pyplot API
plt.hist(student_height);
```

```python
# OO API

fig, ax  = plt.subplots()
ax.hist(student_height, bins = 30);
```

```python
# Subplots
#Option 1: 
x = np.linspace(1, 10, 100)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols = 2, figsize = (10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x ** 2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(drink_prices.keys(), drink_prices.values());
ax4.hist(student_height, bins = 30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                   ncols = 2,
                   figsize = (10, 5))
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x ** 2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(drink_prices.keys(), drink_prices.values());
ax[1,1].hist(student_height, bins = 30);
```
