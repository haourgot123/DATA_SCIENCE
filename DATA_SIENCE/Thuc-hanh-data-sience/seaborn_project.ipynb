```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
cereal_df = pd.read_csv('cereal.csv')
```

```python
cereal_df.head()
```

```python
# irrelevent fields
fields = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis = 1)
```

```python
cereal_df_new.head()
```

Corr tim ra moi tuong quan giua cac gia tri so voi nhau


```python
# Corr()
cereal_corr = cereal_df_new.corr(numeric_only = True) 
cereal_corr
```

```python
# ones_like can buil a matrix of Boolean with same shape as our of data
ones_corr = np.ones_like(cereal_corr, dtype = 'bool')
ones_corr
```

```python
# np.triu : return  only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data = cereal_corr, mask = mask);
```

```python
adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
adjusted_cereal_corr
```

```python
fig, ax = plt.subplots(figsize=(10,8))


#That method uses HUSL colors, so you need hue, saturation, and lightness. 
#I used hsluv.org to select the colors of this chart.
cmap = sns.diverging_palette(0, 230, 90, as_cmap = True)
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap,
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);
ax.set(title = 'CEREAL')
```
