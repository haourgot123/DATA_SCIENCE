Goal: Predict the sales price of each house

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train.csv', index_col='Id')
```

```python
data.head()
```

```python
data.columns
```

# FEATURE SELECTION 
Chose features to train ML

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath','BedroomAbvGr','TotRmsAbvGrd']
```

# SPLITTING DATASET

```python
X = data[features]
```

```python
X
```

```python
X.info()
```

```python
Y = data['SalePrice']
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, train_size= 0.8,test_size=0.2, random_state = 0)
```

```python
X_train
```

# TRAINING MACHINE LEARNING MODEL

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
#fit training data into model

dt_model.fit(X_train, Y_train)
```

```python
y_predic = dt_model.predict(X_val.head())
```

```python
y_predic
```

```python
pd.DataFrame({'y' : Y_val.head(),
             'y_predict' :  y_predic})
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state= 1)
rf_model.fit(X_train, Y_train)
```

```python
y_predics = rf_model.predict(X_val)
```

```python
y_predics[:5]
```

```python
pd.DataFrame({'y' : Y_val,
              'y_predic' : y_predics}).head()
```
